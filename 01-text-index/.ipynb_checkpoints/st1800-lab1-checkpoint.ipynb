```python
input_file = open(path_in+"0704.3504.txt", "r", encoding='iso-8859-1')
output_file_filtered = open(path_out+"0704.3504_filtered.txt", "w")
output_file_drop = open(path_out+"0704.3504_drop.txt", "w")
filecontent = input_file.read()
```

```python
import matplotlib.pyplot as plt

from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import WordNetLemmatizer

porter = PorterStemmer()
lancaster=LancasterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()

filecontent = re.sub('[^A-Za-z0-9]+',' ',filecontent)
tokens = nltk.word_tokenize(filecontent)
# Remove stop words
tokens = [w1 for w1 in tokens if w1 not in stop_words]
tokens = [w1 for w1 in tokens if len(w1)>1]
tokens = [w1.lower() for w1 in tokens]

tokens = [porter.stem(w1) for w1 in tokens]

#tokens = [wordnet_lemmatizer.lemmatize(w1, pos="v") for w1 in tokens ]
tokens = [wordnet_lemmatizer.lemmatize(w1) for w1 in tokens ]

#fdist1 = nltk.FreqDist(filecontent.split(' '))
fdist1 = nltk.FreqDist(tokens)
print(len(fdist1))
#mostcommon = fdist1.most_common(len(fdist1))
mostcommon = fdist1.most_common(20)
x,y = zip(*mostcommon)
plt.figure(figsize=(15,10))
plt.barh(x,y)
plt.xticks(rotation=90)
plt.show()
```

```python
from operator import itemgetter
newfd = sorted(fdist1, key=fdist1.get, reverse=True)[:10]
newfd
fdist1
```

```python
for line in input_file:
    salida_filtered = ""
    salida_drop = ""
    line = re.sub('[^A-Za-z0-9]+',' ',line)
    palabras = line.split()
    for w in palabras:
#        w = re.sub('[^A-Za-z0-9]+','',w)
        w = w.lower()
        if (len(w) >1) and w.isalpha():
#        if (len(w) >1) and w.isalpha() and (w in words.words()) and (w not in stop_words):
#        if (w in words.words()) and (w in stop_words):
            salida_filtered=salida_filtered+" "+w
        else:
            salida_drop=salida_drop+" "+w                
    if (salida_filtered!=""):
        salida_filtered=salida_filtered+"\n"
        output_file_filtered.write(salida_filtered)
        
    if (salida_drop!=""):
        salida_drop=salida_drop+"\n"    
        output_file_drop.write(salida_drop)
```

```python
output_file_filtered.close()
output_file_drop.close()
```

```python
input_file_filtered = open(path_out+"0704.3504_filtered.txt", "r", encoding='iso-8859-1')
```

```python
d = {}
for line in input_file_filtered:
    words = line.split()
    for w in words:
        d[w] = d.get(w, 0) + 1
```

```python
word_freq = []
for key, value in d.items():
    if (key not in stop_words):
            word_freq.append((key, value))
```

```python
from operator import itemgetter
word_freq.sort(key=itemgetter(1),reverse=True) 
```

```python
import csv

with open(path_out+'0704.3504_tf.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(["word", "frecuency"])
    writer.writerows(word_freq)

csvFile.close()
```

```python
# extract top 30 words
top_words = word_freq[:20]
print(top_words)
```
